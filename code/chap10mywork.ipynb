```python
def SamplingDistributions(live, iters=101):
        t = []
        for _ in range(iters):
            sample = thinkstats2.ResampleRows(live)
            ages = sample.agepreg
            weights = sample.totalwgt_lb
            estimates = thinkstats2.LeastSquares(ages, weights)
            t.append(estimates)
        inters, slopes = zip(*t)
        return inters, slopes
```

```python
def Summarize(estimates, actual=None):
        mean = thinkstats2.Mean(estimates)
        stderr = thinkstats2.Std(estimates, mu=actual)
        cdf = thinkstats2.Cdf(estimates)
        ci = cdf.ConfidenceInterval(90)
        print('mean, SE, CI', mean, stderr, ci)
```

```python
actual_inter, actual_slope = thinkstats2.LeastSquares(live.agepreg,
                                                      live.totalwgt_lb)
```

```python
estimated_inters, estimated_slopes = SamplingDistributions(live)
```

```python
Summarize(estimated_inters, actual_inter)
```

```python
Summarize(estimated_slopes, actual_slope)
```

```python
def PlotConfidenceIntervals(xs, inters, slopes, percent=90, **options):
    fys_seq = []
    for inter, slope in zip(inters, slopes):
        fxs, fys = thinkstats2.FitLine(xs, inter, slope)
        fys_seq.append(fys)
        
    p = (100 - percent) / 2
    percents = p, 100 - p
    low, high = thinkstats2.PercentileRows(fys_seq, percents)
    thinkplot.FillBetween(fxs, low, high, **options)
```

```python
PlotConfidenceIntervals(live.agepreg, estimated_inters, estimated_slopes,
                        percent=90)
PlotConfidenceIntervals(live.agepreg, estimated_inters, estimated_slopes,
                        percent=50)
```
