Note that in the CDF plot below:
- I'm calculating the exact y values - i.e., the formula is for the CDF of an exponential distribution, which means I actually use thinkplot.Plot and not thinkplot.Cdf.
- The x axis seems wrong - it's using, I think, the position in the list as the x value and not the actual value for x. Put differently, the list ranges from 0-99 I think while the values actually used range from 0 to about three...?
- I think the analytic.py file has code that does this correctly - see MakeExpoCdf.

```python
def expCdf(lam):
    return [(1 - e**(-lam*x)) for x in linspace(0,3,100)]
```

```python
thinkplot.Plot(expCdf(0.5),label='lam = %.1f' % 0.5)
thinkplot.Plot(expCdf(1),label='lam = %.1f' % 1)
thinkplot.Plot(expCdf(2),label='lam = %.1f' % 2)
thinkplot.Plot(expCdf(3),label='lam = %.1f' % 3)
thinkplot.Plot(expCdf(5),label='lam = %.1f' % 5)

thinkplot.Show(xlabel='some value',ylabel='CDF',axis=[0, 100, 0, 1])
```
