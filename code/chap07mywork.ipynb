```python
import brfss
import thinkstats2
import thinkplot
```

```python
df = brfss.ReadBrfss(nrows=None)
len(df)
```

```python
sample = thinkstats2.SampleRows(df, 5000)
heights, weights = sample.htm3, sample.wtkg2
```

```python
thinkplot.Scatter(heights, weights)
thinkplot.Show(xlabel='Height (cm)',
               ylabel='Weight (kg)',
               axis=[140, 210, 20, 200])
```

```python
heights_jitter = thinkstats2.Jitter(heights, 1.3)
weights_jitter = thinkstats2.Jitter(weights, 0.5)
```

```python
thinkplot.Scatter(heights_jitter, weights_jitter)
thinkplot.Show(xlabel='Height (cm)',
               ylabel='Weight (kg)',
               axis=[140, 210, 20, 200])
```

```python
thinkplot.Scatter(heights_jitter, weights_jitter, alpha=0.1)
thinkplot.Show(xlabel='Height (cm)',
               ylabel='Weight (kg)',
               axis=[140, 210, 20, 200])
```

```python
df_drop = df.dropna(subset=['htm3','wtkg2'])
bins = np.arange(135, 210, 5)
indices = np.digitize(df_drop.htm3, bins)
groups = df_drop.groupby(indices)
```

```python
heights = [group.htm3.mean() for i, group in groups]
cdfs = [thinkstats2.Cdf(group.wtkg2) for i, group in groups]
```

```python
for percent in [75, 50, 25]:
    weights = [cdf.Percentile(percent) for cdf in cdfs]
    label = '%dth' % percent
    thinkplot.Plot(heights, weights, label=label)
thinkplot.Show(xlabel='height (cm)', ylabel='weight (kg)')
```
